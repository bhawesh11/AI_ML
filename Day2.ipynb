```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a,b = 5,10
print("Sum :", a+b)

numbers = [1,2,3,4,5]
for n in numbers:
  if n%2 != 0:
    print(n**2)


arr = np.array([1,4,9,16,25])
print("Mean: ",np.mean(arr))
print("Median: ",np.median(arr))
print("Standard Deviation: ",np.std(arr))


data = {
    "Name": ["Amit", "Neha", "Raj", "Simran"],
    "Age": [25, 28, 22, 30],
    "Score": [88, 92, 95, 70]
}

df = pd.DataFrame(data)
display(df)

print(df.describe())
print("Mean score:", df["Score"].mean())
df[df["Score"] > 90]

df["Result"] = ["Pass" if score >= 90 else "Fail" for score in df["Score"]]
display(df)


plt.plot(df["Age"], df["Score"], marker='o', linestyle='-')
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score")
plt.grid(True)
plt.show()

plt.bar(df["Name"], df["Score"])
plt.xlabel("Name")
plt.ylabel("Score")
plt.title("Name vs Score")
plt.show()

sns.scatterplot(data=df, x="Age", y="Score")
plt.title("Age vs Score")
plt.show()


data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Study_Hours": [2, 3, 4, 1, 5, 3, 6, 4, 2, 5],
    "Marks": [60, 65, 70, 50, 85, 68, 90, 75, 55, 88]
}
df = pd.DataFrame(data)
display(df)

print("Average Marks:", df["Marks"].mean())
print(df[df["Marks"] > df["Marks"].mean()])
sns.scatterplot(x="Study_Hours", y="Marks", data=df)
plt.title("Study Hours vs Marks")
plt.show()

df["Efficiency"] = df["Marks"] / df["Study_Hours"]
display(df)
sns.barplot(x="Name", y="Efficiency", data=df)
plt.title("Efficiency of Students")
plt.show()
sns.regplot(x="Study_Hours", y="Marks", data=df)
plt.title("Regression Plot")
plt.show()
```
